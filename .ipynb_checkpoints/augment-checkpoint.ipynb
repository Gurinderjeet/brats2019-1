```python
import numpy as np
import dataloader
import augment
from matplotlib import pyplot as plt
import math
%matplotlib inline
```

```python
train_set = dataloader.SimpleReader(path='/home/localadmin/segthor/data/train_resampled2_2_25',patch_size=(128,128,128), multiplier=500, patches_from_single_image=2)
```

```python
batch = train_set[0][0][0].numpy()
print(batch[0][0].min(),batch[0][0].max())
```

```python
plt.imshow(batch[0,64,:,:])
plt.show()
```

```python
image = batch[0]
transformation = augment.create_identity_transformation(image.shape)
```

```python
transformation += augment.create_elastic_transformation(
        image.shape,
        control_point_spacing = 20,
        jitter_sigma = 2)
```

```python
transformation += augment.create_rotation_transformation(
        image.shape,
        0.5)
```

```python
image = augment.apply_transformation(image, transformation)
```

```python
plt.imshow(image[64,:,:])
plt.show()
```
